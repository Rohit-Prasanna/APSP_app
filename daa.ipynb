a)Insertion sort

```python
def insertion_srt(a):
  for i in range (1,len(a)):
    j=i
    while a[j-1]>a[j] and j>0:
      a[j-1],a[j]=a[j],a[j-1]
      j-=1

arr = [12, 11, 13, 5, 6]
insertion_srt(arr)
print(arr)
```

b)Bubble sort

```python
def bubble_sort(a):
  for i in range(len(a)-1,0,-1):
    for j in range(i):
      if a[j]>a[j+1]:
        a[j],a[j+1]=a[j+1],a[j]
arr = [64, 34, 25, 12, 22, 11, 90]
bubble_sort(arr)
print(arr)
```

c)Selection sort


```python
def selection_sort(a):
  for i in range (0,len(a)-1):
    min_i=i
    for j in range(i+1,len(a)):
      if a[j]<a[min_i]:
        min_i=j
    a[i],a[min_i]=a[min_i],a[i]

arr = [64, 34, 25, 12, 22, 11, 90]
selection_sort(arr)
print(arr)
```

d)Merge sort

```python
def merge_sort(a):
  if len(a)>1:
    mid =len(a)//2
    left=a[:mid]
    right=a[mid:]
    merge_sort(left)
    merge_sort(right)
    i=j=k=0
    while i< len(left) and j<len(right):
      if left[i]<right[j]:
        a[k]=left[i]
        i+=1
      else:
        a[k]= right[j]
        j+=1
      k+=1
    while i<len(left):
      a[k]=left[i]
      i+=1
      k+=1
    while j<len(right):
      a[k]=right[j]
      j+=1
      k+=1
    return a

arr=[64, 34, 25, 12, 22, 11, 90]
print(merge_sort(arr))
```

e)Quick sort


```python
def quick_sort(a,left,right):
  if left < right:
    partition_pos= partition(a,left,right)
    quick_sort(a,left,partition_pos-1)
    quick_sort(a,partition_pos+1,right)
def partition(a,left,right):
  i=left
  j= right-1
  pivot=a[right]
  while i<j:
    while i<right and a[i]<pivot:
      i+=1
    while j>=left and a[j]>pivot:
      j-=1
    if i<j:
      a[i],a[j]=a[j],a[i]
  if a[i]>pivot:
    a[i],a[right]=a[right],a[i]
  return i

arr=[64, 34, 25, 12, 22, 11, 90]
quick_sort(arr,0,len(arr)-1)
print(arr)
```

f)Heap_sort

```python
def heap(arr, n, i):
    largest = i
    l = 2 * i + 1
    r = 2 * i + 2
    if l < n and arr[i] < arr[l]:
        largest = l
    if r < n and arr[largest] < arr[r]:
        largest = r
    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        heap(arr, n, largest)

def heapSort(arr):
    n = len(arr)
    for i in range(n // 2 - 1, -1, -1):
        heap(arr, n, i)
    for i in range(n - 1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]
        heap(arr, i, 0)

arr = [7, 18, 45, 1, 99, 33]
heapSort(arr)

print('Sorted array is:')
for i in arr:
    print(i)
```

g)Bucket sort

```python
def bucket_sort(arr):
    max_val = max(arr)
    max_digits = len(str(max_val))
    print("Max digits:", max_digits)

    for digit_place in range(max_digits):
        buckets = [[] for _ in range(10)]
        for num in arr:
            digit = (num // (10 ** digit_place)) % 10
            buckets[digit].append(num)
        arr = [num for bucket in buckets for num in bucket]
        print(f"Pass {digit_place + 1}: {arr}")

    return arr

arr = [129, 7, 22, 37, 18, 19, 46, 51, 42, 45]
sorted_arr = bucket_sort(arr)
print("\nSorted array:", sorted_arr)
```

```python
def bucket_sort(arr):
    if len(arr) <= 1:
        return

    max_value = max(arr)
    num_digits = len(str(max_value))
    exp = 1

    while exp <= max_value:
        sort(arr, exp)
        exp *= 10

def sort(arr, exp):
    # Create buckets for each digit (0-9)
    buckets = [[] for _ in range(10)]

    # Place each number in its corresponding bucket
    for number in arr:
        bucket_index = (number // exp) % 10
        buckets[bucket_index].append(number)

    index = 0
    # Concatenate the numbers from the buckets back into the array
    for bucket in buckets:
        for number in bucket:
            arr[index] = number
            index += 1

# Test the bucket sort
array = [15, 1, 321, 10, 802, 2, 123, 90, 109, 11]
bucket_sort(array)
print("Sorted array is:")
print(array)
```

2)BFS and DFS

```python
def bfs(graph,startnode):
  visited=[]
  queue=[startnode]
  while queue:
    curr_node=queue.pop(0)
    visited.append(curr_node)
    for i in graph[curr_node]:
      if i not in visited:
        queue.append(i)
  return visited
graph ={'A':['B','C'],
        'B':['D','E'],
        'C':['F','G'],
        'D':[],
        'E':['H','I'],
        'F':[],'G':[],
        'H':[],'I':[]}
print(bfs(graph,'A'))

```

```python
def dfs(graph,startnode):
  visited=[]
  stack=[startnode]
  while stack:
    curr_node=stack.pop()
    visited.append(curr_node)
    for i in graph[curr_node]:
      if i not in visited:
        stack.append(i)
  return visited

graph={'A':['B','C'],
       'B':['D','E'],
       'C':['F','G'],
       'D':[],
       'E':['H','I'],
       'F':[],'G':[],
       'H':[],'I':[]}
print(dfs(graph,'A'))
```

b)krukal

```python
class Graph:
    def __init__(self, vertices):
        self.V = vertices
        self.graph = []

    def addEdge(self, u, v, w):
        self.graph.append([u, v, w])

    def find(self, parent, i):
        if parent[i] != i:
            parent[i] = self.find(parent, parent[i])
        return parent[i]

    def union(self, parent, rank, x, y):
        if rank[x] < rank[y]:
            parent[x] = y
        elif rank[x] > rank[y]:
            parent[y] = x
        else:
            parent[y] = x
            rank[x] += 1

    def KruskalMST(self):
        result = []
        i = 0
        e = 0
        self.graph = sorted(self.graph, key=lambda item: item[2])
        parent = []
        rank = []
        for node in range(self.V):
            parent.append(node)
            rank.append(0)
        while e < self.V - 1:
            u, v, w = self.graph[i]
            i = i + 1
            x = self.find(parent, u)
            y = self.find(parent, v)

            if x != y:
                e = e + 1
                result.append([u, v, w])
                self.union(parent, rank, x, y)
        minimumCost = 0
        print("Edges in the constructed MST")
        for u, v, weight in result:
            minimumCost += weight
            print("%d -- %d == %d" % (u, v, weight))
        print("Minimum Spanning Tree cost:", minimumCost)

# Driver code
if __name__ == "__main__":
    g = Graph(4)
    g.addEdge(0, 1, 10)
    g.addEdge(0, 2, 6)
    g.addEdge(0, 3, 5)
    g.addEdge(1, 3, 15)
    g.addEdge(2, 3, 4)

    g.KruskalMST()
```

```python
def kruskal(graph):
    parent = {}
    def find(node):
        if parent[node] == node:
            return node
        return find(parent[node])

    def union(node1, node2):
        root1 = find(node1)
        root2 = find(node2)
        parent[root1] = root2

    edges = []
    for node in graph:
        for neighbor, cost in graph[node].items():
            edges.append((cost, node, neighbor))
    edges = sorted(edges)

    for node in graph:
        parent[node] = node

    mst = []
    for edge in edges:
        cost, node1, node2 = edge
        if find(node1) != find(node2):
            union(node1, node2)
            mst.append(edge)
    return mst
```

a)bellman algorithm

```python
def bellman_ford(graph, V, src):
    dist = [float('inf')] * V
    dist[src] = 0

    for _ in range(V - 1):
        for u, v, w in graph:
            if dist[u] != float('inf') and dist[u] + w < dist[v]:
                dist[v] = dist[u] + w

    for u, v, w in graph:
        if dist[u] != float('inf') and dist[u] + w < dist[v]:
            print("Graph contains negative weight cycle")
            return

    return dist
```

a)floyd algorithm

```python
def floyd_warshall(graph):
    V = len(graph)
    dist = list(map(lambda i: list(map(lambda j: j, i)), graph))

    for k in range(V):
        for i in range(V):
            for j in range(V):
                dist[i][j] = min(dist[i][j], dist[i][k] + dist[k][j])

    return dist
```

5)Binary search algorithm

```python
def binary_search(arr,x):
  left=0
  right=len(arr)-1
  while left<=right:
    mid=(left+right)//2
    if arr[mid]==x:
      return mid
    if arr[mid]>x:
      right=mid-1
    else:
      left=mid+1
  return -1

arr=[1,2,3,4,5,6,7,8,9]
x=5
print(binary_search(arr,x))
```

6)Fibonacci Series

```python
def fibo_dp(n):
  if n==0 or n==1:
    return n
  dp=[0]*(n+1)
  dp[0]=0
  dp[1]=1

  for i in range(2,n+1):
    dp[i]=dp[i-1]+dp[i-2]
  return dp[n]

print(fibo_dp(5))
```

7)Max sub array-Kadanes algorithm

```python
def kadane(a):
  max_g=max_c=a[0]
  for i in range (1,len(a)):
    max_c=max(a[i],max_c+a[i])
    if max_c>max_g:
      max_g=max_c
  return max_g

arr=[-2, -3, 4, -1, -2, 1, 5, -3]
print(kadane(arr))
```

```python
def job_seq(arr,t):
  n=len(arr)
  arr.sort(key=lambda x:x[2],reverse=True)
  result=[False]*t
  job=['-1']*t
  for i in range(n):
    for j in range(min(t-1,arr[i][1]-1),-1,-1):
      if result[j]==False:
        result[j]=True
        job[j]=arr[i][0]
        break
  print(job)

arr=[['a',2,100],['b',1,19],['c',2,27],['d',1,25],['e',3,15]]
print(job_seq(arr,3))
```

9)Knapsack algorithm

```python
def knapsack(W,wt,val,n):
  if n==0 or W==0:
    return 0
  if wt[n-1]>W:
    return knapsack(W,wt,val,n-1)
  else:
    return max(val[n-1]+knapsack(W-wt[n-1],wt,val,n-1),knapsack(W,wt,val,n-1))

val=[60,100,120]
wt=[10,20,30]
W=50
n=len(val)
print(knapsack(W,wt,val,n))
```

10)multiple chain multipication

```python
def mcm(d):
  n = len(d)-1
  m = {(i,j):float('inf')for i in range(1,n+1)for j in range(i,n+1)}
  for i in range(1,n+1):
    m[(i,i)]=0
  for i in range(1,n+1):
    for j in range(i,n+1):
      if i==j:
        m[(i,j)]=0
      else:
        m[(i,j)]=min(m[(i,k)]+m[(k+1,j)]+d[i-1]*d[k]*d[j] for k in range(i,j))
  return m[(1,n)]

d=[1,2,3]
print(mcm(d))
```

11)Longest Common subsequence

```python
def get_lcs_length(S1, S2):
    m = len(S1)
    n = len(S2)
    dp = [[0] * (n + 1) for _ in range(m + 1)]

    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if S1[i - 1] == S2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1] + 1
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])

    return dp[m][n]

S1 = "rohit"
S2 = "prasanna"
print("Length of LCS is", get_lcs_length(S1, S2))
```
